# Linear regression

Linear regression is a supervised learning method used to model the relationship between a dependent variable y and one or more independent variables. Model estimates the slope and intercept of the line of best fit 

### Assumption of LR

1. Linearity: The relationship between features and the target is linear in parameters
2. Independence of Errors: Residuals must be independent of each other.
3. No multicollinearity: Independent variables should not be highly correlated with each other.
4. Normality of Errors: The residuals should be normally distributed.


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Generate synthetic data for linear regression
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)


X
```

```python
class LinearRegression:
    def __init__(self, lr, epochs):
        self.w = None
        self.b = None
        self.lr = lr
        self.epochs = epochs

    def predict(self,X):
        return np.dot(X, self.w) + self.b
    
    def fit(self,X,y):
        n_samples = X.shape[0]
        n_features = X.shape[1]
        self.w = np.zeros((n_features,1))
        self.b = 0.0

        for _ in range(self.epochs):
            y_pred = self.predict(X)

            dw = (1/n_samples)*np.dot(X.T, (y_pred - y))
            db = (1/n_samples)*np.sum(y_pred - y)

            self.w -= self.lr * dw
            self.b -= self.lr * db
```

```python
model = LinearRegression(lr=0.01, epochs=1000)
model.fit(X, y)
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='True data')
plt.plot(X, y_pred, color='red', linewidth=2, label='Predicted line')
```
